# 6. 10 Publisher dengan Penjualan Terlaris

```python
best_sell_glb = df.groupby(['Publisher'])['Global_Sales'].agg('sum').reset_index(name='TotalSales')
best_sell_glb = best_sell_glb.sort_values(by='TotalSales', ascending=False).head(11)
best_sell_glb
```

### 7.  Jumlah Penjualan Global Pertahun

```python
best_sell_glb = df.groupby(['Year'])['Global_Sales'].agg('sum').reset_index(name='TotalSales')
best_sell_glb = best_sell_glb.sort_values(by='Year', ascending=True)
best_sell_glb
```

### 8.  10 Frekuensi Tertinggi Tahunan di Dalam Data/Database 

```python
year_counts = df['Year'].value_counts().reset_index()
year_counts.columns = ['Year', 'Frequency']
year_counts['Percent'] = (year_counts['Frequency'] / year_counts['Frequency'].sum()) * 100
freq_year = year_counts.sort_values(by='Frequency', ascending=False).head(11)
freq_year
```

```python
plt.figure(figsize=(20, 6))
plt.bar(freq_year['Year'], freq_year['Frequency'])
plt.title('10 Frekuensi Tertinggi Tahunan')
```

### 9. 10 Platforms Paling Laris di Dunia

```python
best_sell_plforms = df.groupby(['Platform'])['Global_Sales'].agg('sum').reset_index(name='TotalSales')
best_sell_plforms = best_sell_plforms.sort_values(by='TotalSales', ascending=False).head(11)
best_sell_plforms
```

```python
plt.bar(best_sell_plforms['Platform'], best_sell_plforms['TotalSales'])
plt.title('10 Platforms Paling Laris di Dunia')
```
